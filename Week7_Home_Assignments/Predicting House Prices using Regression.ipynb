```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
original_dataset = pd.read_csv(r"D:\AI Engineer Training\Python_Coding\Week7_Home_Assignments\house_price_regression_dataset.csv")
```

```python
plt.scatter(original_dataset['Square_Footage'], original_dataset['House_Price'])
```

```python
x = original_dataset[['Square_Footage']]
y = original_dataset['House_Price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2) 
```

```python
type(x_test)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_new = {'Feature': [6,7]}
y_new_dataframe = pd.dataframe(y_new)
```

```python
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R^2 Score:", r2_score(y_test, y_pred))
```

```python
plt.scatter(x_test, y_test, color='blue', label='Actual')
plt.plot(x_test, y_pred, color='red', label='Predicted')
plt.xlabel('Square Footage')
plt.ylabel('House Price')
plt.title('Actual vs Predicted House Prices')
```
